## Challenge 1:

<div class="alert alert-block alert-danger"> <b>CHALLENGE!</b> Write a function <code>binary_to_decimal()</code> that converts an <b>unsigned</b> binary string, <code>binary_str</code>, to its decimal equivalent without using built-in conversion functions.

```python
# solution
def binary_to_decimal(binary_str):
    decimal_num = 0
    for i in range(len(binary_str) - 1, -1, -1):
        if binary_str[i] == '1':
            decimal_num += 2**(len(binary_str) - 1 - i)
    # # also works:
    # decimal_num = 0
    # for i in range(len(binary_str)):
    #     if binary_str[len(binary_str)-1-i] == '1':
    #         decimal_num += 2**i
    # # also works:
    # return sum([2**(len(binary_str)-1-i) if d=='1' else 0 for i,d in enumerate(binary_str)])
    return decimal_num

result = binary_to_decimal("101010")
print(result)  # Output: 42
```

## Challenge 2:

<div class="alert alert-block alert-danger"> <b>CHALLENGE!</b> Write a function <code>decimal_to_binary()</code> that converts a decimal number, <code>decimal_num</code>, to its <b>unsigned</b> binary representation without using built-in conversion functions.


```python
#solution1
def decimal_to_binary(decimal_num):
    if decimal_num == 0:
        return "0"
    binary_str = ""
    while decimal_num > 0:
        remainder = decimal_num % 2
        binary_str = str(remainder) + binary_str
        decimal_num //= 2
    return binary_str

result = decimal_to_binary(42)
print(result)  # Output: "101010"
```

```python
#solution2
def decimal_to_binary(decimal_num):
    if decimal_num == 0:
        return "0"
    import numpy as np
    power = int(np.floor(np.log(decimal_num)/np.log(2)))+1
    binary_str = ""
    for i in range(power):
        if decimal_num >= 2**(power-1-i):
          binary_str += "1"
          decimal_num -= 2**(power-1-i)
        else:
          binary_str += "0"
    return binary_str

result = decimal_to_binary(42)
print(result)  # Output: "101010"
```

## Challenge 3:

<div class="alert alert-block alert-danger"> <b>CHALLENGE!</b> Write a Python function `binary_subtraction()` that performs binary subtraction on two <b>unsigned</b> binary strings and returns the result as an <b>unsigned</b> binary string. The function should handle binary numbers of different lengths.  Assume that the first number is greater than or equal to the second number (the result will be positive).



```python
#solution1 (use previously defined functions to convert to decimal,
            # subtract as decimals, then convert back to binary)
def binary_subtraction(bin_str1, bin_str2):
    dec_str1 = binary_to_decimal(bin_str1)
    dec_str2 = binary_to_decimal(bin_str2)
    dec_results = dec_str1 - dec_str2
    result = decimal_to_binary(dec_results)
    return result

result = binary_subtraction("1101", "1010")
print(result)  # Output: "11"
```

```python
#solution2 (subtraction by borrowing, directly in binary representation)
def binary_subtraction(bin_str1, bin_str2):
    max_len = max(len(bin_str1), len(bin_str2))
    bin_str1 = bin_str1.zfill(max_len)
    bin_str2 = bin_str2.zfill(max_len)

    borrow = 0
    result = ""

    for i in range(max_len - 1, -1, -1):
        bit1 = int(bin_str1[i])
        bit2 = int(bin_str2[i])

        # Apply borrow if necessary
        bit1 -= borrow

        if bit1 < bit2:
            borrow = 1
            bit1 += 2
        else:
            borrow = 0

        result = str(bit1 - bit2) + result

    # Remove leading zeros
    result = result.lstrip('0')

    if result == "":
        return "0"

    return result

result = binary_subtraction("1101", "1010")
print(result)  # Output: "11"
```
